# Level 3 ‚Äì Task 1: Rating Text Sentiment Analysis

## üîç Objective:
Use the `Rating text` column as a proxy for customer sentiment and analyze its relationship with:
1. Aggregate rating scores
2. Number of votes

## üõ†Ô∏è Approach:
- Clean and filter the `Rating text` column
- Categorize ratings into Positive, Neutral, Negative
- Analyze how sentiment affects overall rating and popularity (votes)

## üìà Insight:
This analysis simulates customer sentiment using available labels and connects them to numerical metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Dataset.csv")

# Clean: Drop missing 'Rating text'
df = df.dropna(subset=['Rating text', 'Aggregate rating', 'Votes'])

# Map rating text to sentiment
sentiment_map = {
    'Excellent': 'Positive',
    'Very Good': 'Positive',
    'Good': 'Neutral',
    'Average': 'Neutral',
    'Poor': 'Negative',
    'Not rated': 'Unknown'
}

df['Sentiment'] = df['Rating text'].map(sentiment_map)

# Filter out 'Unknown'
df = df[df['Sentiment'] != 'Unknown']

# Plot 1: Sentiment vs Aggregate Rating
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Sentiment', y='Aggregate rating', hue='Sentiment', palette='Set2', legend=False)
plt.title("Aggregate Rating by Sentiment Category")
plt.tight_layout()
plt.show()

# Plot 2: Sentiment vs Votes
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Sentiment', y='Votes', hue='Sentiment', palette='Set3', legend=False)
plt.yscale('log')  # votes can be skewed; log scale helps
plt.title("Number of Votes by Sentiment Category")
plt.tight_layout()
plt.show()
```
