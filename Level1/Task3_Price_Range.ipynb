# Level 1 ‚Äì Task 3: Price Range Distribution

## üîç Objective:
Analyze the pricing structure of restaurants in the dataset by:
1. Creating a histogram or bar chart to visualize the distribution of `Price range`.
2. Calculating the percentage of restaurants in each price range category.

## üõ†Ô∏è Approach:
- Inspect and clean the `Price range` column.
- Count restaurants by price range.
- Calculate the % distribution for each category.
- Visualize using a bar plot.

## üìà Insights:
This task provides clarity on how restaurants are priced in the market‚Äîhighlighting whether low-cost or premium dining dominates.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Dataset.csv")

# Check unique price ranges
print("Unique Price Ranges:", df['Price range'].unique())

# Drop nulls if any
df = df.dropna(subset=['Price range'])

# Count restaurants in each price range
price_counts = df['Price range'].value_counts().sort_index()
price_df = price_counts.reset_index()
price_df.columns = ['Price Range', 'Restaurant Count']

# Calculate % of restaurants per category
total = price_df['Restaurant Count'].sum()
price_df['Percentage'] = (price_df['Restaurant Count'] / total) * 100

# Output stats
print(price_df)

# Plot
plt.figure(figsize=(7,5))
sns.barplot(data=price_df, x='Price Range', y='Percentage', hue='Price Range', palette='coolwarm', legend=False)
plt.title("Distribution of Restaurants by Price Range")
plt.ylabel("Percentage (%)")
plt.xlabel("Price Range")
plt.tight_layout()
plt.show()
```
