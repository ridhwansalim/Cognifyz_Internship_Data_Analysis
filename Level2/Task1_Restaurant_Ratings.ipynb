# Level 2 ‚Äì Task 1: Restaurant Ratings

## üîç Objective:
1. Analyze the distribution of `Aggregate rating` across restaurants.
2. Determine the most common rating range.
3. Calculate the average number of votes received by restaurants.

## üõ†Ô∏è Approach:
- Clean the relevant columns (`Aggregate rating`, `Votes`)
- Plot a histogram or bar chart of ratings
- Count and highlight the most frequent rating range
- Calculate average votes

## üìà Insights:
Understanding rating distribution reveals how critical users are and how competitive restaurants need to be to stand out.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Dataset.csv")

# Clean: drop rows with missing ratings or votes
df = df.dropna(subset=['Aggregate rating', 'Votes'])

# Convert to numeric
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

# Rating distribution
rating_counts = df['Aggregate rating'].value_counts().sort_index()

# Most common rating
most_common_rating = rating_counts.idxmax()
print(f"‚≠ê Most common rating: {most_common_rating} stars")

# Average votes
avg_votes = df['Votes'].mean()
print(f"üìä Average number of votes per restaurant: {avg_votes:.2f}")

# Plot distribution (no warning version)
rating_df = pd.DataFrame({'Rating': rating_counts.index, 'Count': rating_counts.values})
plt.figure(figsize=(8,5))
sns.barplot(data=rating_df, x='Rating', y='Count', hue='Rating', palette='rocket', legend=False)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()
```
